```python
import bs4 as bs #Biblioteca para web scrapping
import urllib.request #Trabalhar com URL (caminhos de sites)
import nltk #Trabalhar com PLN
import spacy #Trabalhar com PLN
```

```python
pln = spacy.load('pt_core_news_sm')
pln
```

```python
type(pln('estou com fome'))
```

```python
for tokens in pln('computador está quebrado'):
  print(tokens.text, tokens.pos_)
```

```python
for tokens in pln('estou estar estarei'):
  print(tokens.text, tokens.lemma_)
```

```python
dados = urllib.request.urlopen('https://pt.wikipedia.org/wiki/Intelig%C3%AAncia_artificial')
```

```python
dados = dados.read()
dados
```

```python
dados_html = bs.BeautifulSoup(dados, 'lxml')
dados_html
paragrafos = dados_html.find_all('p')
conteudo = ''
for p in paragrafos:
  conteudo += p.text
```

```python
doc = pln(conteudo)
```

```python
for entidade in doc.ents:
  print(entidade.text, entidade.label_)
```

```python
from spacy import displacy
displacy.render(doc, style = 'ent', jupyter = True)
```
